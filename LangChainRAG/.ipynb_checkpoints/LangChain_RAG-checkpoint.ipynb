```python
import os
import sys
from pathlib import Path

from dotenv import load_dotenv

#LangChain document loaders
from langchain_community.document_loaders import (
    TextLoader,
    PyMuPDFLoader,
    Docx2txtLoader,
    UnstructuredHTMLLoader,
    UnstructuredMarkdownLoader,
    CSVLoader,
)

#LangChain splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter

#LangChain embeddings + vector store
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_chroma import Chroma

#LangChain LCEL pipeline 
from langchain_classic.chains import RetrievalQA
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
```

```python
load_dotenv()

DOCS_DIR        = "./docs"
CHROMA_DIR      = "./chroma_db"
COLLECTION_NAME = "rag_demo"
EMBED_MODEL     = "text-embedding-3-small"
CHAT_MODEL      = "gpt-4o-mini"
CHUNK_SIZE      = 500          # tokens / characters per chunk
CHUNK_OVERLAP   = 80           # overlap between consecutive chunks
TOP_K           = 4            # number of chunks to retrieve per query
```

```python
LOADER_MAP = {
    ".txt":  TextLoader,
    ".pdf":  PyMuPDFLoader,
    ".docx": Docx2txtLoader,
    ".html": UnstructuredHTMLLoader,
    ".htm":  UnstructuredHTMLLoader,
    ".md":   UnstructuredMarkdownLoader,
    ".csv":  CSVLoader,
}
```

```python
def load_documents(docs_dir: str):
    """Load all supported documents from docs_dir."""
    docs_path = Path(docs_dir)
    if not docs_path.exists():
        raise FileNotFoundError(f"Docs directory not found: {docs_dir}")

    documents = []
    skipped   = []

    for filepath in sorted(docs_path.iterdir()):
        ext = filepath.suffix.lower()
        if ext not in LOADER_MAP:
            skipped.append(filepath.name)
            continue

        loader_cls = LOADER_MAP[ext]
        print(f"  file Loading [{ext}]: {filepath.name}")
        try:
            loader = loader_cls(str(filepath))
            docs   = loader.load()
            # Ensure every doc carries its source filename in metadata
            for doc in docs:
                doc.metadata.setdefault("source", filepath.name)
            documents.extend(docs)
        except Exception as e:
            print(f"  Could not load {filepath.name}: {e}")

    if skipped:
        print(f"   Skipped (unsupported): {', '.join(skipped)}")

    return documents
```

```python
def chunk(documents):
    """Split documents into overlapping chunks."""
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=CHUNK_SIZE,
        chunk_overlap=CHUNK_OVERLAP,
        separators=["\n\n", "\n", ". ", " ", ""],  # tries paragraph → sentence → word
    )
    return splitter.split_documents(documents)
```

```python
def ingest(directory,embed):
    documents = load_documents(directory)
    chunks = chunk(documents)
    vectorstore = Chroma.from_documents(
        documents=chunks,
        embedding=embed,
        collection_name=COLLECTION_NAME,
        persist_directory=CHROMA_DIR,
    )
    print("chroma created")
```

```python
def chat_chain(query,retriever,llm,Type):
    qa_stuff = RetrievalQA.from_chain_type(
    llm=llm, 
    chain_type=Type, 
    retriever=retriever, 
    verbose=True
    )
    return qa_stuff.run(query)
```

```python
def format_docs(docs):
    """Join retrieved docs into a single context string."""
    return "\n\n".join(
        f"[Source: {d.metadata.get('source', 'unknown')}]\n{d.page_content}"
        for d in docs
    )
PROMPT = ChatPromptTemplate.from_messages([
    ("system",
     "You are a helpful assistant. Use ONLY the context below to answer the "
     "question. If the answer is not in the context, say: "
     "'I don't have information about that in my knowledge base.' "
     "Always mention which source document(s) your answer came from.\n\n"
     "Context:\n{context}"),
    ("human", "{question}"),
])
# LCEL chain: retrieve → format → prompt → LLM → parse
rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough(),
    }
    | PROMPT
    | llm
    | StrOutputParser()
)

# We also need the raw docs for displaying sources, so build a parallel chain
retrieve_chain = (
    RunnablePassthrough()
    | retriever  # returns List[Document]
)
```

```python
embeddings = OpenAIEmbeddings(model=EMBED_MODEL)

vectorstore = Chroma(
    collection_name=COLLECTION_NAME,
    embedding_function=embeddings,
    persist_directory=CHROMA_DIR,
)
llm = ChatOpenAI(model=CHAT_MODEL, temperature=0.2, max_tokens=512)
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": TOP_K},
)
```

```python
ingest(DOCS_DIR,embeddings)
```

```python
chat_chain("hello",retriever,llm,"stuff")
```

```python
while True:
    try:
        query = input("You> ").strip()
    except (KeyboardInterrupt, EOFError):
        print("\n\n👋  Goodbye!")
        break

    if not query:
        continue
    if query.lower() in {"exit", "quit"}:
        print("\n👋  Goodbye!")
        break

    print("\n🔍  Retrieving & generating ...\n")

    try:
        # Run retrieval and generation in parallel so we have both docs and answer
        source_docs = retrieve_chain.invoke(query)
        answer      = rag_chain.invoke(query)
    except Exception as e:
        print(f"❌  Error: {e}\n")
        continue

    # ── Show sources ──
    print("📚  Sources used:")
    seen = set()
    for i, doc in enumerate(source_docs, 1):
        src = doc.metadata.get("source", "unknown")
        preview = doc.page_content[:80].replace("\n", " ")
        if src not in seen:
            print(f"  [{i}] {src}")
            seen.add(src)
        print(f"      \"{preview}{'...' if len(doc.page_content) > 80 else ''}\"")

    # ── Print answer ──
    print("Agent>")
    print(answer)
    print()
    print()
```
